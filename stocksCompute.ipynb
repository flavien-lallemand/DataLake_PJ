```
#Create a DataFrame with data from DB
query = "(Select * FROM FL.StocksData) sparkAlias"
df = spark.read.jdbc(url=jdbcUrl, table=query, properties=connectionProperties)
df.show()
```

```
#Compute Daily Return Rate : return a df with computed daily return rate
def daily_return_rate(stock_name, start_date, end_date):
    df_company = df.filter(df.company_name == stock_name)
    df_date = df_company.filter(df_company.Date >= start_date).filter(df_company.Date <= end_date)
    df_date = df_date.withColumn("daily_return_rate", df_date["Close"] - df_date["Open"]-1 / df_date["Open"])
    
    return df_date[['Date', 'company_name', 'daily_return_rate']]
```

```
from pyspark.sql.window import Window
from pyspark.sql.functions import avg, col, lag, sum

#Compute Moving Average : return a df with computed daily return rate
def moving_average(stock_name, start_date, end_date, nb_moving_points):
    df_company = df.filter(df.company_name == stock_name)
    df_date = df_company.filter(df_company.Date >= start_date).filter(df_company.Date <= end_date)
    window_spec = Window.partitionBy("company_name").orderBy("Date").rowsBetween(-(nb_moving_points-1), 0)
    df_date = df_date.withColumn("moving_average", avg(col("Open")).over(window_spec))

    
    return df_date[['Date', 'company_name', 'moving_average']]
```

```
df_moving_average = moving_average("TESLA", "2019-01-01", "2019-01-31", 5)
df_moving_average.show()

#Compute for APPLE
df_moving_average_APPLE = moving_average("APPLE", "2020-06-01", "2020-06-31", 5)
upload_to_blob("moving_average_APPLE_1_MONTH.csv", df_daily_return_APPLE)
df_daily_return_APPLE.show()

#Compute for GOOGLE
df_moving_average_GOOGLE = moving_average("GOOGLE", "2019-12-01", "2020-02-31", 6)
upload_to_blob("moving_average_GOOGLE_3_MONTHS.csv", df_daily_return_GOOGLE)
```
